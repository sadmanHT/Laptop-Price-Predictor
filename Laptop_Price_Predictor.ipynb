```python
#Here we are breaking the ScreenResolution into two parts
new = laptop_df["ScreenResolution"].str.split("x", n = 1, expand = True)
```

```python
laptop_df["X_resolution"] = new[0]
laptop_df["Y_resolution"] = new[1]
laptop_df.sample(5)
```

```python
#fixing the x resolution
laptop_df["X_resolution"] = laptop_df["X_resolution"].str.replace(",","").str.findall(r'(\d\.?\d+)').apply(lambda x:x[0])
```

```python
laptop_df["X_resolution"] = laptop_df["X_resolution"].astype("int")
laptop_df["Y_resolution"] = laptop_df["Y_resolution"].astype("int")
```

```python
laptop_df.corr()["Price"]
```

Calculating PPI(Pixels Per Inches)
Formula: ((resolution of x)^2 + (resolution of y)^2)/Inches

```python
laptop_df["PPI"] = (((laptop_df["X_resolution"]**2) + (laptop_df["Y_resolution"]**2))**0.5).astype("float")
laptop_df.head()
```

```python
laptop_df.drop("X_resolution",inplace = True, axis = 1)
laptop_df.drop("Y_resolution",inplace = True, axis = 1)
```

```python
laptop_df.drop("ScreenResolution", axis = 1, inplace = True)
```

```python
laptop_df["CPU Name"] = laptop_df["Cpu"].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
laptop_df.head()
```

```python
def fetch_processor(text):
  if text == "Intel Core i7" or text == "Intel Core i5" or text == "Intel Core i3":
    return text
  else:
    if text.split()[0] == "Intel":
      return "Other Intel Processor"
    else:
      return "AMD Processor"
```

```python
laptop_df["CPU Brand"] = laptop_df["CPU Name"].apply(fetch_processor)
laptop_df.head()
```

```python
laptop_df["CPU Brand"].value_counts().plot(kind = "bar")
```

```python
sns.barplot(x = laptop_df["CPU Brand"], y = laptop_df["Price"])
plt.xticks(rotation = "vertical")
```
